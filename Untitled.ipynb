```python
import pandas as pd
import numpy as np
```

```python
grades = pd.read_csv("student_grades.csv")

print("Student Gradebook")
print(grades)
```

```python
print("\nNumber of Students :", len(grades))
print("Number of Assignments :", len(grades.columns)-1)

print("\nColumn Names")
print(grades.columns.tolist())
```

```python
assignment_columns = grades.columns[1:]

grades["Average"] = grades[assignment_columns].mean(axis=1)
grades["Median"] = grades[assignment_columns].median(axis=1)
grades["Highest Score"] = grades[assignment_columns].max(axis=1)
grades["Lowest Score"] = grades[assignment_columns].min(axis=1)

print("\nStudent Performance")
print(grades)
```

```python
print("\nAssignment Statistics")

for assignment in assignment_columns:
    print(f"\n{assignment}")
    print("Average :", grades[assignment].mean())
    print("Median  :", grades[assignment].median())
    print("Highest :", grades[assignment].max())
    print("Lowest  :", grades[assignment].min())
```

```python
class_average = grades["Average"].mean()

print("Overall Class Average =", round(class_average,2))
```

```python
sorted_grades = grades.sort_values(by="Average", ascending=False)

print("Students Ranked by Average Score\n")

print(sorted_grades[["Name","Average"]])
```

```python
top_student = sorted_grades.iloc[0]

print("🏆 Top Performer")

print("Name :", top_student["Name"])
print("Average :", round(top_student["Average"],2))
```

```python
grades["Result"] = np.where(grades["Average"]>=40,"Pass","Fail")

print(grades[["Name","Average","Result"]])
```

```python
def calculate_grade(avg):

    if avg>=90:
        return "A+"

    elif avg>=80:
        return "A"

    elif avg>=70:
        return "B"

    elif avg>=60:
        return "C"

    elif avg>=50:
        return "D"

    else:
        return "F"

grades["Grade"] = grades["Average"].apply(calculate_grade)

print(grades[["Name","Average","Grade"]])
```

```python
scores = grades[assignment_columns].values

print("Overall Average :", np.mean(scores))
print("Overall Median :", np.median(scores))
print("Highest Score :", np.max(scores))
print("Lowest Score :", np.min(scores))
print("Standard Deviation :", round(np.std(scores),2))
```

```python
grades.to_csv("Student_Report.csv", index=False)

print("Report saved successfully.")
```
